```python
import numpy as np 
import pandas as pd 
import librosa 
import os 
```

```python
def extract_features(file_path):
    audio , sr = librosa.load(file_path , sr =22050)
    mfcc = np.mean(librosa.feature.mfcc(y=audio , sr=sr , n_mfcc = 40).T , axis=0)
    chroma = np.mean(librosa.feature.chroma_shft( y = audio , sr=sr).T , axis=0)
    mel = np.mean(librosa.feature.melspectrogram(y=audio , sr=sr).T , axis=0)
    zcr = np.mean(librosa.feature.zero_crossing_rate(y=audio , sr=sr).T , axis = 0)
    return np.hstack([mfcc , chroma , mel , zcr])
```
